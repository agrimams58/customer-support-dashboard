```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 50)
pd.set_option('display.width', 200)

df = pd.read_csv('../data/raw/customer_support_tickets.csv')
print(f"Loaded {len(df):,} rows and {df.shape[1]} columns")
```

```python
print("COLUMNS:")
print(df.columns.tolist())
print("\nDTYPES:")
print(df.dtypes)
print("\nFIRST 3 ROWS:")
df.head(3)

```

```python
missing = df.isna().sum()
missing_pct = (missing / len(df) * 100).round(1)
pd.DataFrame({'missing': missing, 'pct': missing_pct}).sort_values('missing', ascending=False)
```

```python
cat_cols_to_check = ['Ticket Type', 'Ticket Priority', 'Ticket Channel', 'Ticket Status']

for col in cat_cols_to_check:
    if col in df.columns:
        print(f"\n--- {col} ---")
        print(df[col].value_counts(dropna=False))
```

```python
date_like = [c for c in df.columns if any(k in c.lower() for k in ['date', 'time', 'response', 'resolution'])]
print("Date-like columns:", date_like)

for col in date_like:
    print(f"\n{col}: sample values →", df[col].dropna().head(3).tolist())
    print(f"  → non-null count: {df[col].notna().sum():,}")
```

```python
if 'Customer Satisfaction Rating' in df.columns:
    csat = df['Customer Satisfaction Rating']
    print(f"CSAT populated: {csat.notna().sum():,} of {len(df):,} ({csat.notna().mean()*100:.1f}%)")
    print(f"\nCSAT distribution:")
    print(csat.value_counts(dropna=False).sort_index())
```
